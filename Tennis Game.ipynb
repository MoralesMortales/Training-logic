```python
""" /*
 * Escribe un programa que muestre cómo transcurre un juego de tenis y quién lo ha ganado.
 * El programa recibirá una secuencia formada por "P1" (Player 1) o "P2" (Player 2), según quien
 * gane cada punto del juego.
 * 
 * - Las puntuaciones de un juego son "Love" (cero), 15, 30, 40, "Deuce" (empate), ventaja.
 * - Ante la secuencia [P1, P1, P2, P2, P1, P2, P1, P1], el programa mostraría lo siguiente:
 *   15 - Love
 *   30 - Love
 *   30 - 15
 *   30 - 30
 *   40 - 30
 *   Deuce
 *   Ventaja P1
 *   Ha ganado el P1
 * - Si quieres, puedes controlar errores en la entrada de datos.   
 * - Consulta las reglas del juego si tienes dudas sobre el sistema de puntos.   
 */ """
 
print("Welcome to game system (Tennis edition)")

""" Here is a class """

class player:
    name = ""
    points = 0

    def __init__(self, myName) -> None:
        name = myName 
    
""" int Main(){ """

p1 = player(input("What's the name of player one or their team?"))
p2 = player(input("What's the name of player one or their team?"))

rounds = 0
maxPoints = 40

print("\nStart\n")

while p2.points < maxPoints and p1.points < maxPoints:
    winningPlayer = str(input("Who won the last point? (P1 - P2)")).upper()
    
    while winningPlayer != "P1" and winningPlayer != "P2":
        print("Opcion no admitida")
        winningPlayer = str(input("Who won the last point? (P1 - P2)")).upper()
    
    winner = player(winningPlayer == "P1" if p1 else p2) 
    loser = player(winningPlayer == "P1" if p2 else p1)
    
    rounds = rounds + 1
    print(f"\n<<n Round {rounds} >>\n")

    winner.points = winner.points + 15 
    
    if winner.points > 40:
        winner.points = 40
    
    if loser.points == winner.points:
        print(f"Deuce -- {winner.name} has currently {winner.points} ----- {loser.name} has currently {loser.points}")
    elif loser.points == 0:
        print(f"P1 has currently {winner.points} ----- {loser.name} is in LOVE")
    else:      
        print(f"P1 has currently {winner.points} ----- {loser.name} has currently {loser.points}")

    
p1.points > p2.points if print(f"\nHa ganado {p1.name}") else print(f"\nHa ganado {p2.name}")
```
